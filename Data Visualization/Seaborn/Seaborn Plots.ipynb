## Seaborn Plots

| Plot | Function | Primary Purpose | Best Used For | Data Type |
|------|----------|----------------|--------------|-----------|
| Line Plot | `sns.lineplot()` | Show trends over time | Time-series analysis, trends | Numerical |
| Scatter Plot | `sns.scatterplot()` | Show relationship between variables | Correlation, feature relationships | Numerical |
| Bar Plot | `sns.barplot()` | Compare categories | Category comparison with averages | Categorical + Numerical |
| Count Plot | `sns.countplot()` | Count occurrences | Frequency analysis | Categorical |
| Histogram | `sns.histplot()` | Show data distribution | Understanding spread and shape | Numerical |
| KDE Plot | `sns.kdeplot()` | Density estimation | Smooth distribution visualization | Numerical |
| Box Plot | `sns.boxplot()` | Detect outliers | Outlier detection, spread analysis | Numerical |
| Violin Plot | `sns.violinplot()` | Distribution + density | Compare distributions across groups | Numerical |
| Strip Plot | `sns.stripplot()` | Show individual observations | Small datasets, point distribution | Numerical + Categorical |
| Swarm Plot | `sns.swarmplot()` | Non-overlapping observations | Detailed categorical analysis | Numerical + Categorical |
| Heatmap | `sns.heatmap()` | Visualize matrix/correlation | Correlation analysis, feature selection | Matrix/Correlation |
| Pair Plot | `sns.pairplot()` | Relationships among features | Exploratory Data Analysis (EDA) | Multiple Numerical Features |
| Joint Plot | `sns.jointplot()` | Scatter + distributions | Bivariate analysis | Numerical |
| Regression Plot | `sns.regplot()` | Regression visualization | Trend and prediction analysis | Numerical |
| Cat Plot | `sns.catplot()` | Flexible categorical plots | Comparing categorical variables | Categorical |

### Quick Selection Guide

| Question | Recommended Plot |
|-----------|------------------|
| How is data distributed? | `histplot()`, `kdeplot()` |
| Are there any outliers? | `boxplot()` |
| How are two variables related? | `scatterplot()` |
| Is there a trend over time? | `lineplot()` |
| Which features are correlated? | `heatmap()` |
| Compare categories? | `barplot()` |
| Count category occurrences? | `countplot()` |
| Visualize all feature relationships? | `pairplot()` |
| Need a regression line? | `regplot()` |
| Compare distributions by category? | `violinplot()` |

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
```

```python
sns.set_theme(style='darkgrid')
```

```python
tips
```

```python
tips.to_csv('tips.csv')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
```

```python
# 1. scatter plot
sns.scatterplot(data=tips,x="total_bill", y="tip",hue="time",size="size", palette="deep")
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
# 2. Line Plots
sns.lineplot(data=tips,x="size", y="tip",hue='sex')
plt.title("Lineplot of Size vs Tip")
plt.show()
```

```python
sns.lineplot(data=tips, x= 'size', y='total_bill', hue='sex',ci=None, markers='o')
plt.title("Lineplot of Total Bill vs Size")
plt.show()
```

```python
tips.columns
```

```python
#3. bar plot

sns.barplot(data=tips, x='day', y='total_bill', hue = 'sex',palette='muted')

plt.title("Barplot of Total Bill by Day")
plt.show()
```

```python
# 4. Boxplot 

sns.boxplot(data=tips, x='day', y='tip', hue='smoker', palette='Set2')
plt.title("Boxplot of Tips by Day and Smoker Status")
plt.show()
```

```python
# 5. violin plot 

sns.violinplot(data=tips, x='day', y='total_bill', hue='time', split=True, palette='pastel')
plt.title("Violin Plot of Total Bill by Day and Time")  
plt.show()
```

```python
# 6. count plot 

sns.countplot(data=tips, x='day', hue='smoker', palette='dark')
plt.title("Count Plot of Days by Smoker Status")
plt.show()
```

```python
#7. regression plot 

sns.regplot(data=tips, x='total_bill', y='tip',scatter_kws={'s': 50, 'alpha': 0.6, 'color': 'green'},line_kws={'color':'red'})
plt.title("Regression Plot of Total Bill vs Tip")   
plt.show()
```

```python
# 8. Histogram of total bill with KDE

sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='blue')
plt.title("Histogram of Total Bill with KDE")
plt.show()
```

```python
# 9. pairplot

sns.pairplot(tips, hue='sex', vars=["total_bill", "tip", "size"], palette='husl')
plt.suptitle("pair plot: numerberic variables by gender", y=1.02)
plt.show()
```

```python
# 10 catplot 

sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='point', palette='bright')
plt.title("catplot(point):Tips by day and gender")
plt.show()
```

```python
# 11. jointplot

sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker', color='purple', palette='coolwarm')
plt.suptitle("Jointplot: Total Bill vs Tip", y=1.02)
plt.show()
```

```python
sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.histplot, 'total_bill', bins=20, kde=True).add_legend()
plt.show()
```

```python
# 13. strip plot

sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True, palette='Set1')
plt.title("strip plot: Tips by data and gender")
plt.show()
```

```python
# 14. KDE PLOT

sns.kdeplot(data=tips, x='total_bill',hue='sex', fill=True, palette='tab10')
plt.title("kde plot:Total bill density by gender")
plt.show()
```
