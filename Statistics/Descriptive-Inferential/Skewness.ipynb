## Skewness

```python
# Incomes of data
l_data = [20,25,22,28,24,26,23,27,500,1000]
```

```python
# Order the data in ascending order
l_data.sort()
l_data
```

```python
# Find the mean
mean_incomes = sum(l_data)/len(l_data)
print(f"Mean of the incomes {mean_incomes}")
```

```python
# Find the median
n = len(l_data)
if n%2==0:
    median_incomes=(l_data[(n//2)-1]+l_data[((n)//2)])/2
else:
    median_incomes=l_data[((n//2))]

print(f"Median of the incomes {median_incomes}")
```

```python
# Other way to find the median using pandas
import pandas as pd
incomes = pd.Series(l_data)
median_incomes=incomes.median()

print(f"Median of the incomes {median_incomes}")
```

```python
mode_incomes = incomes.mode()

print(f"Mode of the incomes {mode_incomes}")
```

```python
mean_incomes = incomes.mean()
print(f"Median of the incomes {median_incomes}")
print(f"Mode of the incomes {mode_incomes}")
print(f"Mean of the incomes {mean_incomes}")
```

```python
# Plot the data skewness
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(incomes,kde=True)
plt.title("Incomes Distribution")
plt.xlabel("Incomes")
plt.ylabel("Frequency")
```
