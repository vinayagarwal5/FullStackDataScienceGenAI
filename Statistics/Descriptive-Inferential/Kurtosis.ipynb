## Kurtosis

```python
list = [2,4,4,4,5,5,7,9]
```

```python
import numpy as np
data = np.array(list)
print(f"Data {data}")
```

```python
# Mean for the data
mean = np.mean(data)
```

```python
# Deviations from the mean
deviation = data - mean

# Median for the data
l = len(data)
if l%2==0:
    median= (data[((l//2)-1)+(l//2)])/2
else:
    median= data[((l//2)-1)]
    
median
```

```python
# 4 power deviations
dev_4power = deviation ** 4
```

```python
# kurtosis calculation
kurtosis = (sum(dev_4power) / len(data)) / (np.std(data) ** 4)
kurtosis
```

```python
# Find the which type
if (kurtosis < 3):
    print("Platty")
elif (kurtosis > 3) :
    print("Lepto")
else:
    print("Meso")
    
```

```python
# Plot the data skewness and kurtosis
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(data,kde=True)
plt.title("Data Distribution")
plt.xlabel("Data")
plt.ylabel("Frequency")
```

```python
# Plot Skewness
import statistics
median = statistics.median(data)
mode = statistics.mode(data)

print(f"Mean = {mean} and Median = {median} and Mode = {mode}")

plt.figure(figsize=(8, 5))
sns.kdeplot(data)
plt.axvline(mean, label="Mean")
plt.axvline(median, color="black", label="Median")
plt.axvline(mode.squeeze(), color="green", label="Mode")
plt.title("Distribution of Data (Skewness)")
plt.xlabel("Data")
plt.legend()
plt.show()
```

```python
# Kurtosis

plt.figure(figsize=(8, 5))
sns.kdeplot(data, color="red")
plt.title("KDE Plot Showing Kurtosis of Diamond Prices")
plt.xlabel("Data")
plt.ylabel("Density")
plt.show()
```
