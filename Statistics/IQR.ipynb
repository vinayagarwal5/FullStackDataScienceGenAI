```python
data = [7,15,36,39,40,41,42,43,47]
```

```python
Q1= data[len(data)//4]

Q3= data[len(data)*3//4]
print(Q1)
print(Q3)
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print(lower_bound)
print(upper_bound)
```
