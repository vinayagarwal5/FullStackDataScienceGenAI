## Onehot Encoding

```python
data = read_csv('encoding.csv')
```

```python
categorical_columns = data.select_dtypes(include='object').columns
```

```python
# import the label encoder
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
OHE = OneHotEncoder()
#apply label encoder on categorical columns

encoded_data = OHE.fit_transform(data[categorical_columns])

encoded_data
encoded_df = pd.DataFrame(encoded_data.toarray(),columns=OHE.get_feature_names_out(categorical_columns))
```

```python
encoded_df
```

```python
# convert into integer type
encoded_df = encoded_df.astype(int)
encoded_df
```

```python
# concatinate the encoded data frame 

# Before concatinate the data drop the categorical columns sinc we had new encoded columns
df.drop(columns=categorical_columns)
```

```python
Final_DF = pd.concat([df,encoded_df],axis=1)
```
