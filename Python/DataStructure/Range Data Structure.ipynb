# 🔢 Python Range Data Structure

The `range` data type is an immutable sequence used to generate a series of numbers.

It is commonly used in:
- ✔ Loops  
- ✔ Iteration  
- ✔ Index generation  

---

# 📌 Syntax

```python
range(start, stop, step)

<table style="width:80%; font-size:14px; border-collapse:collapse; margin:auto;"> <thead> <tr style="background-color:#f2f2f2;"> <th style="color:purple; border:1px solid #333; padding:10px;">Parameter</th> <th style="color:purple; border:1px solid #333; padding:10px;">Description</th> </tr> </thead> <tbody> <tr> <td style="border:1px solid #333; padding:8px;"><b>start</b></td> <td style="border:1px solid #333; padding:8px;"> Starting value (inclusive) </td> </tr> <tr> <td style="border:1px solid #333; padding:8px;"><b>stop</b></td> <td style="border:1px solid #333; padding:8px;"> Ending value (exclusive) </td> </tr> <tr> <td style="border:1px solid #333; padding:8px;"><b>step</b></td> <td style="border:1px solid #333; padding:8px;"> Increment or decrement value </td> </tr> </tbody> </table>

````python
range(5)
````

````python
list(range(5))
````

````python
list(range(0,5,2))
````

````python
data_range = list(range(0,15,3))
print(data_range)
````

````python
data_range[1]
````

````python
# List operations can be performed since it's in the list
data_range[1] =4
print(data_range)
````

````python
range(5,40,2)
````

````python
tuple(range(5,40,2))
````

````python
set(range(5,40,2))
````

````python
s = {range(5,40,2)}
````

````python
for i in range(5,10,2):
    print(i)
````
