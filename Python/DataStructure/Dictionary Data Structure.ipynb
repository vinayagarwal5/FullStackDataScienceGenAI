# 📘 Features of Python Dictionary

A Python dictionary is a powerful and flexible data structure used to store data in **key-value pairs**.

````python
person = {
    "name": "Roja",
    "age": 22,
    "course": "Python"
}

print(person)
````

````python
print(f"Person Name {person['name']}")
print(f"Person Age {person['age']}")
print(f"Person course {person['course']}")
````

````python
print(f"Keys : {person.keys()}")
print(f"Values : {person.values()}")
````

````python
# Get value based on key if exists return value else won't return anything
person.get('name')
````

````python
# List of items
person.items()
````

````python
# Pop is based on key
person.pop('name')
````

````python
print(person)
````

````python
# Pop is based on LIFO
person = {
    "name": "Roja",
    "age": 33,
    "course": "Python",
    "guest": "Avinar",
    "special": "Charvik",
}
person.popitem()
````

````python
# Update the dictionary
person.update(city="Hyderabad")
person.update({"data": "Python"})
print(f"Updated Data: {person}")
````

````python
# Update the same keys dictionary
person.update(city="Hyderabad1")
person.update({"data": "Python1"})
print(f"Updated Data: {person}")
````

````python
person['data']='Test'
print(f"Updated Data: {person}")
````

````python
# Adding the new keys dictionary
person['test']='Python'
print(f"Updated Data: {person}")
````

````python
data=person.copy()
print(f"Copied Data {data}")
````

````python
# Adding new item with default None value
data.setdefault('new')
print(f"Copied Data {data}")
````

````python
data.clear()
print(f"Cleared Data {data}")
````

````python
for k,v in person.items():
    print(f"Key {k}: Value {v}")
````

````python
keys = {'a' , 'b' , 'c' , 'd'}
value = {10,20,30}
mydict3 = dict.fromkeys(keys , value) # Create a dictionary from a sequence of
mydict3
````
