<h2 style="color:darkred">Global Functions on LIST</h2>

## Enumarate()

```python
l1 = [1,5,6,8,3]
for i in enumerate(l1):
    #print(i)               # i in interator count
    print(f"Iterator {i[0]}")
    print(f"Value {i[1]}")
```

## All() & Any()

```python
# If 0 in the list all will give False
# If empty list any will give False
l  = []
l1 = [1,5,6,8]
l2 = [True, 'a',1,2]
l3 = [5,0,7]
```

```python
print(all(l))
print(all(l1))
print(all(l2))
print(all(l3))
```

```python
print(any(l))
print(any(l1))
print(any(l2))
print(any(l3))
```

## len()

```python
l2 = [True, 'a',1,2,3+4j]
print(f"Length of the list {len(l2)}")
```

```python
l10 = list()
l11 = []
print(type(l10))
print(type(l11))
```

## ZIP

```python
seq1 = [2,3,4,5,'Roja',1+3j]
seq2 = [2,3,4,5,'Roja1',11+3j]

list(zip(seq1,seq2))
```

```python
seq1 = [2,3,4,5,'Roja',1+3j]
seq2 = [2,3,4,5,'Roja1',11+3j]
seq3 = [1,2]

list(zip(seq1,seq2,seq3))
```

```python
for i,(v,j) in enumerate(zip(seq1,seq2)):
    print(f"Iterator {i} & values {v} and {j}")
```
