#### Conditional Statements

```python
x = 8
r = x%2
if r == 0:
    print('Even Number')

print('Odd Number')
```

```python
x = 8
r = x%2
if r == 0:
    print('Even Number')
else:
    print('Odd Number')
```

```python
x = 9
r = x%2
if r == 0:
    print('Even Number')
else:
    print('Odd Number')
```

```python
x = 9
r = x%2

if r == 0:
    print('Even Number')
elif r == 1:
    print('Odd Number')
else:
    print('Not a number')
```

```python
x = 9
r = x%2

if r == 0:
    print('Even Number')
    if x > 5:
        print("Greater than 5")
elif r == 1:
    print('Odd Number')
    if x > 5:
        print("Greater than 5")
else:
    print('Not a number')
```

### Loops

```python
for i in range(10):
    print(i)
```

```python
fruits = ["Apple", "Banana", "Mango"]

for val in fruits:
    print(val)
```

```python
count = 1
while count < 5 :
    print(count)
    count+= 1
```

```python
for i in range(20):
    if i == 5:
        break;
    print(i)
```

```python
for i in range(20):
    if i == 5:
        continue;
    print(i)
```

```python
for i in range(20):
    pass
```

#### Print Even Numbers till 50

```python
for i in range(50):
    if i%2 == 0:
        print(f"{i} is Even Number")
```

#### Sum Positive Numbers

```python
numbers = [5, -2, 10, -1, 7]
sum_pos = 0
for n in numbers:
    if n > 0:
        sum_pos+= n

print(f"Sum of positive numbers {sum_pos}")
```

#### Loop with else

The else block executes if loop ends naturally.

```python
for i in range(5):
    print(i)
else:
    print("Loop Ends")
```

#### else Does NOT Execute After break

```python
for i in range(5):
    print(i)
    if i==3:
        break
else:
    print("Loop Ends")
```

### Find largest number in a list

```python
l = [2,3,5,819,3,4]
largest = l[0]

for i in l:
    if largest < i:
        largest = i
print(f"Largest number {largest}")
        
```

### Check palindrome number

```python
word = input('Enter a string to check palindrome')
palindrome=''
length = len(word)-1

for i in range(len(word)):
    palindrome+= word[length-i]
if palindrome == word:
    print("Given String is a palindrome")
else:
    print("Given String is not a palindrome")
```

### Count vowels in a string

```python
string = input('Enter a string to count of vowels').lower()
vowels=['a','e','i','o','u']
count = 0

for i in string:
    if i in vowels:
        count+=1

print(f"Count of vowels in {string} is {count}")
```
