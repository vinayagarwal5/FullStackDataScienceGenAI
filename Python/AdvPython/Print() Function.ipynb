# 🖨️ Python print() Function

The `print()` function is used to display output on the screen.

It is one of the most commonly used built-in functions in Python.

---

# 📌 Syntax

```python

````python
print("Avinash","Charvik","Ram",sep = "--@@--",end = "\n") # \n new line
print("Roja")
````
