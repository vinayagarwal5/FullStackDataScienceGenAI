## User-Defined Functions

A **User-Defined Function (UDF)** in Python is a reusable block of code created by the programmer to perform a specific task. Unlike built-in functions such as `print()` or `len()`, UDFs allow you to organize, modularize, and reuse your own logic.

### Key Features

- **Reusable:** Write once, use multiple times.
- **Modular:** Break large programs into smaller, manageable parts.
- **Maintainable:** Easier to update and debug code.
- **Customizable:** Accept inputs (parameters) and return outputs.
- **Improves Readability:** Makes code cleaner and easier to understand.

### Basic Syntax

```python
def function_name(parameters):
    """
    Optional Docstring
    """
    # Function body
    return value
```

### Components of a Function

| Component | Description |
|-----------|-------------|
| `def` | Keyword used to define a function |
| Function Name | Identifier used to call the function |
| Parameters | Inputs passed to the function |
| Function Body | Code executed when the function is called |
| `return` | Sends a value back to the caller |

### Example

```python
def greet(name):
    return f"Hello, {name}!"

message = greet("Alice")
print(message)
```

**Output:**

```text
Hello, Alice!
```

### Types of User-Defined Functions

| Type | Example |
|--------|---------|
| No Arguments, No Return Value | `def hello():` |
| Arguments, No Return Value | `def greet(name):` |
| No Arguments, Return Value | `def get_pi():` |
| Arguments, Return Value | `def add(a, b):` |

### Advantages of UDFs

- Reduces code duplication (**DRY - Don't Repeat Yourself**).
- Improves code organization.
- Simplifies testing and debugging.
- Encourages code reuse.
- Makes programs easier to maintain and scale.

### Best Practices

- Use meaningful function names.
- Keep functions focused on a single task.
- Write docstrings to describe function behavior.
- Avoid making functions unnecessarily long.
- Use parameters instead of hardcoding values.

### Real-World Example

```python
def calculate_area(length, width):
    return length * width

area = calculate_area(10, 5)
print("Area:", area)
```

**Output:**

```text
Area: 50
```

> **Tip:** Functions are one of the most important building blocks in Python and are heavily used in Data Science, Machine Learning, Web Development, and Automation.

## Types of Function Arguments

Python provides several ways to pass data into functions, making them flexible and easy to use.

### 1. Positional Arguments

Arguments are assigned based on their position in the function call.

#### Syntax

```python
def greet(name, age):
    print(name, age)

greet("Alice", 25)
```

**Output:**

```text
Alice 25
```

> The order of arguments must match the order of parameters.

---

### 2. Keyword Arguments

Arguments are passed using parameter names, so the order does not matter.

#### Syntax

```python
def greet(name, age):
    print(name, age)

greet(age=25, name="Alice")
```

**Output:**

```text
Alice 25
```

> Improves readability, especially when functions have many parameters.

---

### 3. Default Arguments

Parameters can have default values that are used when no argument is provided.

#### Syntax

```python
def greet(name, country="India"):
    print(name, "from", country)

greet("Alice")
greet("Bob", "USA")
```

**Output:**

```text
Alice from India
Bob from USA
```

> Default values make parameters optional.

---

### 4. Variable-Length Arguments (`*args`)

Used when the number of positional arguments is unknown.

#### Syntax

```python
def add_numbers(*args):
    print(args)
    print(sum(args))

add_numbers(10, 20, 30, 40)
```

**Output:**

```text
(10, 20, 30, 40)
100
```

> `*args` collects all extra positional arguments into a tuple.

---

### 5. Variable-Length Keyword Arguments (`**kwargs`)

Used when the number of keyword arguments is unknown.

#### Syntax

```python
def student_info(**kwargs):
    print(kwargs)

student_info(name="Alice", age=25, city="Hyderabad")
```

**Output:**

```text
{'name': 'Alice', 'age': 25, 'city': 'Hyderabad'}
```

> `**kwargs` collects all extra keyword arguments into a dictionary.

---

### Summary Table

| Argument Type | Syntax | Description |
|--------------|--------|-------------|
| Positional | `func(10, 20)` | Values assigned by position |
| Keyword | `func(a=10, b=20)` | Values assigned by parameter name |
| Default | `def func(x=10)` | Uses default value if omitted |
| Variable Positional | `*args` | Accepts multiple positional arguments |
| Variable Keyword | `**kwargs` | Accepts multiple keyword arguments |

### Example Combining All Types

```python
def display_info(name, age=18, *args, **kwargs):
    print("Name:", name)
    print("Age:", age)
    print("Args:", args)
    print("Kwargs:", kwargs)

display_info(
    "Alice",
    25,
    "Python",
    "Data Science",
    city="Hyderabad",
    country="India"
)
```

**Output:**

```text
Name: Alice
Age: 25
Args: ('Python', 'Data Science')
Kwargs: {'city': 'Hyderabad', 'country': 'India'}
```

### Best Practice

- Use **positional arguments** for required values.
- Use **keyword arguments** for clarity.
- Use **default arguments** for optional parameters.
- Use `*args` and `**kwargs` when the number of inputs is unknown.

````python
# Defining a function with arguments
def num(a,b):
    print(a)
    print(b)
num(2,3)
````

````python
# Assinging a function with arguments to variable
def num(a,b):
    print(f"Sum {a+b}")
c=num(3,4)
print(c)
````

````python
# Assinging a function with arguments to variable with return
def num(a,b):
    return a+b
c=num(3,4)
print(c)
````

````python
## Positional Arguments
def Employees(eid,name):
    return f"Employee Id {eid} & Employee Name {name}"
employee_details = Employees(1,"Roja")
print(employee_details)
````

````python
## Keyword Arguments
def Employees(eid,name):
    return f"Employee Id {eid} & Employee Name {name}"
employee_details = Employees(eid=1,name="Roja")
print(employee_details)
````

````python
## Default Arguments
def Employees(eid,name,addr="Hyd"):
    return f"Employee Id {eid} & Employee Name {name} & Employee address {addr}"
employee_details = Employees(eid=1,name="Roja")
print(employee_details)
employee_details_blr = Employees(eid=1,name="Roja",addr='blr')
print(employee_details_blr)
````

````python
## Variable Length Arguments
def Employees(eid,name,*data):
    return f"Employee Id {eid} & Employee Name {name} & Employee address {data}"
employee_details = Employees(1,"Roja","Hyd",5000)
print(employee_details)
````

````python
## All types of args
def Employees(eid,*name,**data):
    return f"Employee Id {eid} & Employee Name {name} & Employee address {data}"
employee_details = Employees(1,"Roja","Rani",addr="Hyd",sal=5000)
print(employee_details)
````
