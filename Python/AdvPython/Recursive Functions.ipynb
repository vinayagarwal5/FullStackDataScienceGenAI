## Recursion

**Recursion** is a programming technique in which a function calls itself (directly or indirectly) to solve a problem by breaking it down into smaller, simpler subproblems. Each recursive function must have a **base case** to stop the recursion and prevent infinite function calls.

### Key Characteristics

- **Function calls itself** to solve a smaller version of the same problem.
- **Base case** stops the recursion.
- **Recursive case** continues breaking the problem into smaller subproblems.
- Uses the **call stack**, where each function call is stored until it returns.

### Common Applications

- Mathematical calculations (Factorial, Fibonacci)
- Tree traversal (Preorder, Inorder, Postorder)
- Graph traversal (Depth-First Search - DFS)
- Divide-and-Conquer algorithms (Merge Sort, Quick Sort, Binary Search)
- Backtracking problems (N-Queens, Sudoku Solver, Maze Solving)
- File system and directory traversal

### General Structure

```python
def recursive_function(parameters):
    # Base case
    if condition:
        return result

    # Recursive case
    return recursive_function(smaller_problem)
```

### Example: Factorial

```python
def factorial(n):
    if n == 0 or n == 1:      # Base case
        return 1
    return n * factorial(n - 1)   # Recursive case

print(factorial(5))
```

**Output**

```
120
```

### How Recursion Works

For `factorial(5)`:

```
factorial(5)
= 5 × factorial(4)
= 5 × 4 × factorial(3)
= 5 × 4 × 3 × factorial(2)
= 5 × 4 × 3 × 2 × factorial(1)
= 5 × 4 × 3 × 2 × 1
= 120
```

### Advantages

- Produces clean and concise code.
- Naturally fits hierarchical and recursive data structures.
- Simplifies divide-and-conquer and backtracking algorithms.

### Disadvantages

- Higher memory usage due to the call stack.
- Slower because of function call overhead.
- Deep recursion can cause a `RecursionError` (stack overflow in Python).

## When to Avoid Recursion

Although recursion is elegant and useful for solving many problems, it is not always the best approach. Consider avoiding recursion in the following situations:

- **Simple Repetitive Tasks:** When the problem can be solved easily using loops (`for` or `while`), iteration is often simpler and more efficient.
- **Deep Recursion:** If the recursion depth is very large, it may exceed Python's recursion limit and result in a **`RecursionError` (stack overflow)**.
- **Performance-Critical Applications:** Recursive function calls introduce additional overhead, making them slower than iterative solutions in many cases.

---

## Recursion vs Iteration

| Feature | Recursion | Iteration |
|---------|-----------|-----------|
| **Method** | A function calls itself to repeat the task | Uses loops (`for`, `while`) to repeat steps |
| **Memory Usage** | Higher (each function call uses stack memory) | Lower (no function call stack) |
| **Performance** | Generally slower due to function call overhead | Generally faster and more efficient |
| **Termination** | Stops when the **base case** is reached | Stops when the **loop condition** becomes false |
| **Code Simplicity** | Often shorter and easier for recursive problems | Usually more straightforward for repetitive tasks |
| **Best Used For** | Tree traversal, graph traversal, divide-and-conquer, backtracking | Counting, searching, array processing, repetitive computations |
| **Risk** | May cause `RecursionError` if recursion is too deep | No stack overflow risk from loops |
| **Examples** | Factorial, Fibonacci, Binary Search, DFS | Summing numbers, Traversing arrays, Bubble Sort |

### When to Use Which?

| Use Recursion ✅ | Use Iteration ✅ |
|-----------------|-----------------|
| Tree Traversal | Array Traversal |
| Graph Traversal (DFS) | Counting Loops |
| Binary Search | Searching Lists |
| Merge Sort | Bubble Sort |
| Quick Sort | Selection Sort |
| Backtracking | Repetitive Calculations |
| Divide-and-Conquer Algorithms | Performance-Critical Tasks |

> **Interview Tip:** Prefer **iteration** when efficiency and memory usage are important. Choose **recursion** when the problem has a naturally recursive structure (e.g., trees, graphs, divide-and-conquer algorithms), as it often results in cleaner and more intuitive code.

````python
import sys 
sys.setrecursionlimit(100)
print(sys.getrecursionlimit())

i = 0
def recursive_function():
    global i
    i += 1
    print(i)
    recursive_function()

recursive_function()
````

````python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    
factorial(5)
````

````python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

for i in range(10):
    print(fibonacci(i), end=" ")
````
