# Lambda Functions in Python

A **lambda function** is a small, anonymous function defined without a name using the `lambda` keyword. It is designed to contain **only a single expression**, which is automatically evaluated and returned.

Lambda functions are commonly used for **short, one-time operations**, especially when working with higher-order functions such as `map()`, `filter()`, `sorted()`, and `reduce()`.

---

## Syntax

```python
lambda arguments: expression
```

- **`lambda`** → Keyword used to create an anonymous function.
- **`arguments`** → Zero or more input parameters.
- **`expression`** → A single expression that is automatically returned.

---

## Basic Examples

### 1. Add Two Numbers

```python
add = lambda x, y: x + y

print(add(5, 3))
```

**Output**

```
8
```

---

### 2. Square a Number

```python
square = lambda x: x ** 2

print(square(6))
```

**Output**

```
36
```

---

### 3. Inline If-Else

```python
is_even = lambda x: True if x % 2 == 0 else False

print(is_even(8))
```

**Output**

```
True
```

---

## Common Use Cases

### 1. `map()`

Applies a function to every element in an iterable.

```python
numbers = [1, 2, 3, 4]

squared = list(map(lambda x: x ** 2, numbers))

print(squared)
```

**Output**

```
[1, 4, 9, 16]
```

---

### 2. `filter()`

Filters elements based on a condition.

```python
numbers = [2, 5, 8, 11, 14]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)
```

**Output**

```
[2, 8, 14]
```

---

### 3. `sorted()`

Custom sorting using a key.

```python
students = [
    ("Alice", 25),
    ("Bob", 20),
    ("Charlie", 30)
]

sorted_students = sorted(
    students,
    key=lambda student: student[1]
)

print(sorted_students)
```

**Output**

```
[('Bob', 20), ('Alice', 25), ('Charlie', 30)]
```

---

### 4. `reduce()`

Performs cumulative operations on iterable elements.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

total = reduce(lambda x, y: x + y, numbers)

print(total)
```

**Output**

```
15
```

---

## Lambda vs Regular Function

| Feature | Lambda Function | Regular Function (`def`) |
|----------|-----------------|---------------------------|
| **Name** | Anonymous (no name) | Named function |
| **Syntax** | `lambda args: expression` | `def function():` |
| **Lines of Code** | Single expression only | Multiple statements allowed |
| **Return Statement** | Implicit return | Explicit `return` required |
| **Complex Logic** | Not suitable | Supports loops, conditions, exception handling |
| **Readability** | Best for short operations | Better for complex logic |
| **Reusability** | Usually one-time use | Highly reusable |

---

## When to Use Lambda

✅ Short calculations

✅ One-time functions

✅ `map()`

✅ `filter()`

✅ `sorted()`

✅ `reduce()`

---

## When to Avoid Lambda

❌ Multiple statements are required.

❌ Complex business logic.

❌ Functions requiring loops or exception handling.

❌ When readability becomes difficult.

---

## Quick Comparison

| Task | Lambda Example |
|------|----------------|
| Add numbers | `lambda x, y: x + y` |
| Square a number | `lambda x: x ** 2` |
| Check even | `lambda x: x % 2 == 0` |
| Sort by age | `sorted(data, key=lambda x: x[1])` |
| Filter even numbers | `filter(lambda x: x % 2 == 0, nums)` |
| Square all numbers | `map(lambda x: x**2, nums)` |
| Sum list | `reduce(lambda x, y: x + y, nums)` |

> **Interview Tip:** Use **lambda functions** for concise, one-line operations passed as arguments to higher-order functions. For reusable or complex logic, prefer defining a regular function using `def`.
