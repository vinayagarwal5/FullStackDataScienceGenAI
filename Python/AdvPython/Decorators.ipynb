# Python Decorators

A **decorator** in Python is a function that takes another function as an argument, extends or modifies its behavior **without changing the original function's code**, and returns a new function.

Decorators are applied using the **`@decorator_name`** syntax placed directly above the function definition.

---

## Why Use Decorators?

Decorators help you:

- Reuse code (DRY Principle)
- Add functionality without modifying existing code
- Keep business logic clean and separate
- Execute code before and/or after a function runs

---

## Basic Syntax

```python
def decorator(function):
    def wrapper():
        # Code before function call
        function()
        # Code after function call
    return wrapper

@decorator
def greet():
    print("Hello!")

greet()
```

---

## How It Works

```text
greet()
    │
    ▼
wrapper()
    │
    ├── Before Logic
    │
    ├── greet()
    │
    └── After Logic
```

The decorator wraps the original function, allowing additional code to run before and after it.

---

## Example 1: Basic Decorator

```python
def my_decorator(func):

    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")

    return wrapper


@my_decorator
def say_hello():
    print("Hello!")


say_hello()
```

### Output

```
Before function execution
Hello!
After function execution
```

---

## Example 2: Decorator with Arguments

Use `*args` and `**kwargs` to make the decorator work with any function.

```python
def log_decorator(func):

    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")

        result = func(*args, **kwargs)

        print(f"Result = {result}")

        return result

    return wrapper


@log_decorator
def add(a, b):
    return a + b


add(5, 10)
```

### Output

```
Calling add
Result = 15
```

---

## Preserving Function Metadata

Decorators replace the original function with the wrapper function.

Use **`functools.wraps`** to preserve:

- Function name
- Docstring
- Metadata

```python
from functools import wraps

def decorator(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Before")
        result = func(*args, **kwargs)
        print("After")
        return result

    return wrapper
```

## Common Real-World Use Cases

| Use Case | Description |
|----------|-------------|
| Logging | Log function calls and arguments |
| Authentication | Check user permissions before executing a function |
| Authorization | Restrict access based on roles |
| Performance Monitoring | Measure execution time |
| Caching | Store results of expensive computations |
| Input Validation | Validate function arguments |
| Debugging | Print execution details |
| Retry Mechanism | Retry failed operations automatically |

---

## Built-in Python Decorators

| Decorator | Purpose |
|-----------|---------|
| `@property` | Creates getter methods |
| `@property.setter` | Creates setter methods |
| `@classmethod` | Receives the class (`cls`) instead of an object |
| `@staticmethod` | Doesn't use instance (`self`) or class (`cls`) |
| `@functools.lru_cache` | Caches function results for faster execution |

---

## Decorator Execution Flow

```text
Function Call
      │
      ▼
Decorator
      │
      ▼
Wrapper Function
      │
 ┌────┴────┐
 │ Before  │
 └────┬────┘
      ▼
Original Function
      ▼
 ┌────┴────┐
 │  After  │
 └────┬────┘
      ▼
 Return Result
```

---

## Decorator vs Normal Function

| Feature | Normal Function | Decorator |
|---------|-----------------|-----------|
| Purpose | Performs a task | Extends another function |
| Input | Data | Another function |
| Output | Value | Modified function |
| Code Modification | Direct | Without changing original code |
| Reusability | Moderate | High |

---

## Advantages

- Improves code reusability
- Follows the **DRY (Don't Repeat Yourself)** principle
- Separates business logic from cross-cutting concerns
- Makes code cleaner and easier to maintain
- Easy to stack multiple decorators

---

## Limitations

- Can make debugging more difficult
- Adds a small performance overhead
- Deeply nested decorators may reduce readability

---

## Quick Reference

| Task | Decorator |
|------|-----------|
| Add behavior before function | Decorator |
| Add behavior after function | Decorator |
| Preserve function metadata | `@wraps` |
| Cache results | `@lru_cache` |
| Create getter | `@property` |
| Create class method | `@classmethod` |
| Create static method | `@staticmethod` |

> **Interview Tip:** A decorator is a **higher-order function** that accepts another function, adds extra functionality before and/or after its execution, and returns the modified function **without changing the original source code**.
