### 8.  Write a Python program that accepts a string and prints the number of vowels in it.

```python
string = input("Enter string").lower()
vowels = ['a','e','i','o','u']
count = 0

for i in string:
    if i in vowels:
        count+=1
print(f"Number of vowels in the given string {string} is {count}")
```

### 10. How would you check if a string is a palindrome using input()

```python
p_string = input("Enter a palindrom string")
string_r = ''
for i in p_string[::-1]:
    string_r+=i
if p_string == string_r:
    print(f"Given string {p_string} is a palindrome")
else:
    print(f"Given string {p_string} is not a palindrome")
```

### 26. Write a program that accepts a string and counts the occurrence of a particular character.

```python
string = input("Enter a string")
dict_count = {}

for i in string:
    if i in dict_count:
        dict_count[i] = dict_count[i]+1
    else:
        dict_count[i] = 1

for k,v in dict_count.items():
    print(f"{k} in the given string {v}")
```

### 43. Write a program to find the sum of all digits in a string entered by the user.

```python
string = input("Enter a string")
data = re.findall(r'\d', string)
sum_digits = 0
for num in data:
    sum_digits+=int(num)

print(f"Sum of digits in the string {sum_digits}")
```

### 49. How do you check if a string entered by the user contains any punctuation

```python
import string
text = input("Enter a string")
for char in text:
    if char in string.punctuation:
        print(f"String {text} is having punctuations")
        break
else:
    print(f"String {text} is not having punctuations")
```

### 50. Write a program that accepts a sentence and prints the longest word.

```python
text = input("Enter a string").split()
length = 0
word = ''
for words in text:
    if length < len(words):
        length = len(words)
        word = words

print(f"Longest word is {word} with {length}")
```
