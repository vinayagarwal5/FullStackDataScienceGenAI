## Operations

```python
# Creating arrays
a = np.array([1, 2, 3])
print(a)
# Creating arrays with a range of values evenly spaced
b = np.arange(0,10,3)
print(b)
# Creating arrays filled with zeros
c = np.zeros((2,3))
print(c)
# Creating arrays filled with zeros dtype int
d = np.zeros((2,3), dtype=int)
print(d)
# Creating arrays filled with ones
e = np.ones((2,3))
print(e)
# Creating arrays filled with ones dtype int
f = np.ones((2,3), dtype=int)
print(f)
# Creating identity matrix
g = np.eye(4,dtype=int)
print(g)
```

```python
# Array Manipulation

# Reshaping arrays
h = np.array([1, 2, 3, 4, 5, 6])
h_reshaped = h.reshape(2, 3)
print(h_reshaped)

h_reshaped_2 = h.reshape(3, 2)
print(h_reshaped_2)

# Flattening arrays used in CNN maxpooling layers
h_flattened = h_reshaped.flatten()
print(h_flattened)

# raveling arrays
h_raveled = h_reshaped.ravel()
print(h_raveled)
```

```python
# Transposing arrays

i = np.array([[1, 2, 3], [4, 5, 6]])
i_transposed = i.T
print(i_transposed)

#vstack and hstack
j = np.array([1, 2, 3])
k = np.array([4, 5, 6])
j_vstacked = np.vstack((j, k))
print(j_vstacked) 

j_hstacked = np.hstack((j, k))
print(j_hstacked)
```

```python
## Mathematical Operations

l = np.array([1, 2, 3])
m = np.array([4, 5, 6]) 

# Element-wise addition
add = l + m
print(add)

# Element-wise subtraction
subtract = l - m
print(subtract)

# Element-wise multiplication
multiply = l * m
print(multiply) 

# power
power = l ** 2
print(power)

# Square root
sqrt = np.sqrt(l)
print(sqrt)

# dot product
dot_product = np.dot(l, m)
print(dot_product)
```

```python
# Statistical operations

n = np.array([1, 2, 3, 4, 5])   
# Mean
mean = np.mean(n)
print(mean)
# Median
median = np.median(n)
print(median)
# Standard deviation
std_dev = np.std(n)
print(std_dev)
# Variance
variance = np.var(n)
print(variance)

# Minimum and maximum
min_value = np.min(n)
print(min_value)
max_value = np.max(n)
print(max_value)
```

```python
# linear algebra operations 
o = np.array([[1, 2], [3, 4]])
p = np.array([[5, 6], [7, 8]])
# Matrix multiplication
matrix_product = np.dot(o, p)
print(matrix_product)
   
```

```python
# Random Samling
# Random sampling from a normal distribution
random_normal = np.random.normal(loc=0, scale=1, size=(3, 3))
print(random_normal)
```

```python
# seeding random number generator for reproducibility

# Stop to generate the same random numbers every time we run the code
np.random.seed(42) # Seed helps to stop generating new numbers every time we run the code
random_numbers = np.random.rand(5)
print(random_numbers)
```

```python
# linalg operations
array_a = np.array([[1, 2], [3, 4]])
linalg_inv = np.linalg.inv(array_a)
print(linalg_inv)

np.linalg.det(array_a)

np.linalg.eig(array_a)
```
