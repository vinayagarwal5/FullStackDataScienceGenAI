## Boolean masking & Filtering

# 🎭 Boolean Masking and Filtering in NumPy

Boolean masking is a technique used to filter or select elements from a NumPy array based on conditions.

It returns:
- ✔ `True` for matching elements  
- ✔ `False` for non-matching elements  

This is one of the most powerful features in NumPy for data analysis and machine learning.

---

# 📌 What is a Boolean Mask?

A Boolean mask is an array containing only:
- `True`
- `False`

These values are generated by applying conditions to arrays.

---

# 🔹 Example 1: Creating a Boolean Mask

```python id="m5q2xt"
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

mask = arr > 25

print(mask)
