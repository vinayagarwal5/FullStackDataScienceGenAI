<h2 style="color:darkred">2. id() function</h2>

1. The id() function returns a unique id (or) memory address for the specified object.
2. All objects in Python has its own unique id.
3. The id is assigned to the object when it is created.

## 🆔 id() Syntax
```python
id(object)

````python
s = "welcome"
id(s)
````

````python
i = 10
id(i)
````

````python
s1 = "welcome"     # s variable having the same value so s1 share the same id. Helps for memory optimizations
id(s1)
````

````python
b = s              # b variable copy of s so it'll share the same id. It's a shallow copy. If s changes b will change
id(b)
````

````python
s = "welcomee"     # Here s changed so b also changed. 
print(s)
print(b)
print(id(b))
print(id(s))
````

````python
b = "hi"            # If b changes it won't applicable to s. Since b memory address i.e id will change.
print(s)
print(b)
print(id(b))
print(id(s))
````

<h2 style="color:darkred">5. input() function</h2>
In Python, the input() function is a built-in tool that allows a program to receive data from the user via the keyboard. When called, it pauses the program and waits for the user to type something and press Enter.
<h3 style="color:darkred"> Key Features </h3>
<h3 style="color:purple">Returns a String: </h3>Regardless of what the user types (even if it's a number), the input() function always returns the data as a string (str).
<h3 style="color:purple">Optional Prompt: </h3>You can provide a message to guide the user (e.g., "Enter your name: "), or leave it empty to just wait for input.
<h3 style="color:purple">Pauses Execution: </h3>The program will not proceed to the next line until the user presses the Enter key.

````python
x = input("Enter the 1st string")
y = input("Enter the 2nd string")
z = x + ' ' + y
print (z)
````

````python
a = int(input("Enter the 1st number"))
b = int(input("Enter the 2nd number"))
c = a + b
print (f"Total sum of 2 numbers {c}")
````
