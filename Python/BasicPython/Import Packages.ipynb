# 📦 Python import Statement

The `import` statement allows you to access code from other Python files (modules) or libraries.

It helps in:
- ✔ Reusing code  
- ✔ Organizing programs  
- ✔ Accessing built-in and external libraries  

---

# 🔹 Basic Syntax

import module_name

````python
import math
math.sqrt(16)
````

````python
math.floor (4.7)
````

````python
math.ceil (5.9)
````

````python
# help(math)  # To check the documentation of the package we can use help
````

````python
import numpy as np
x = np.nan
math.isnan(x)
````

````python
from math import pow
math.pow (5,2)
````

````python
print(math.floor(4.6))
print(math.ceil(4.6))
print(math.sqrt(4.6))
print(math.pow(4,2))
````

````python
sqrt(4)             # To use directly need to import package from math import *
````

````python
from math import *   # We can directly access the functions and constants without module/alias name
print(sqrt (4))
print(pow (4,2))
print(floor (4.8))
print(ceil (4.8))
````
