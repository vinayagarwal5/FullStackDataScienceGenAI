<h2 style="color:darkred">Comments</h2>

<div>In Python, comments are notes within the source code that the interpreter ignores during execution. They are primarily used to explain logic, improve code readability, and temporarily disable code during debugging.</div>

<div class="container">
    <h3 style="color:purple">Types of Comments : </h3>
    <div class="section">
        <h4 style="color:green">Single-Line Comments</h4>
        <div>
            These start with a hash symbol (<strong>#</strong>) and continue until the end of the line.
            They can be placed on their own line or at the end of a line of code (inline).
        </div>
        <div>
# This is a standalone single-line comment
x = 5  # This is an inline comment
        </div>
    </div>
    <div class="section">
        <h4 style="color:green">Multi-Line Comments</h4>
        <p>
            Python does not have a dedicated syntax for multi-line comments.
            Developers typically use one of the following methods:
        </p>
        <p><strong>1. Consecutive # symbols:</strong> Recommended practice (PEP 8).</p>
        <p>
# This is a multi-line comment
# using multiple hash symbols
# across lines
        </p>
        <p><strong>2. Triple-quoted strings (""" or '''):</strong>  
        Unassigned string literals are ignored by the interpreter and act like comments.</p>
        <div>
"""
This is a multi-line comment
using triple quotes
"""
        </div>
    </div>

</div>

```python
# Single line comment
val1 = 10
```

```python
# This is a multi-line comment
# using multiple hash symbols
# across lines
val1 = 10
```

```python
# Triple quotes can be used for multiline comments
'''
1. Identifiers should not have special characters except _
2. Identifier should not start with digits
3. Identifiers should not use reserved words
'''
val1 = 10
```

```python
""" This is a 
    multi-line comment 
    using triple quotes """
name = "Roja rani"
```
