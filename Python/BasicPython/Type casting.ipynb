<h1>Type Casting</h1>
Convert one data type to another data type

<h2 style="color:darkred">1. Convert all data types to int</h2>

1. In type casting to int we can't pass 2 arguments
2. We can't type cast from complex to int
3. We can't type cast from text string to int
4. We can type cast remaining all to int types

```python
int (3.5)      # Float to int type cast
```

```python
int (True)     # Boolean to int type cast
```

```python
int ('10')     # Number String to int type cast
```

```python
int (3 + 5j)   # Can't convert complex to int
```

```python
int ('10.5')   # Can't convert Text string to int type cast
```

```python
int ('ten')   # Can't convert Text string to int type cast
```

```python
int (10.5,23.5) # can't pass 2 arguments
```

<h2 style="color:darkred">2. Convert all data types to float</h2>

1. In type casting to float we can't pass 2 arguments
2. We can't type cast from complex to float
3. We can't type cast from text string to float
4. We can type cast remaining all to float types

```python
float ( 23 )      # Convert type to float
```

```python
float ( True )    # Convert boolean to float
```

```python
float ( 3 + 5j )  # Can't convert complex to float
```

```python
float ( 2, 3 )    # can't pass 2 parameters to float type
```

```python
float ('23')      # Convert Number string to float
```

```python
float ('23.5')    # Convert Text string to float
```

```python
float ('ten')     # Convert Text string to float
```

<h2 style="color:darkred">3. Convert all data types to boolean</h2>

1. In type casting to float we can't pass 2 arguments or more
2. We can type cast all to boolean types

```python
bool (2)          # convert int to boolean
```

```python
bool (2.5)        # convert float to boolean
```

```python
bool (3+5j)       # convert complex to boolean
```

```python
bool ('')         # convert empty string to boolean
```

```python
bool ('ten')      # convert string to boolean
```

```python
bool ()           # convert empty to boolean
```

```python
bool (None)       # convert none to boolean
```

```python
bool (2,3)        # can't pass 2 parameters to bool type cast
```

<h2 style="color:darkred">4. Convert all data types to complex</h2>

1. In type casting to float we can't pass 3 arguments or more
2. We can't pass 2nd argument as string
3. We can type cast remaining all to complex types

```python
complex (10)       # convert int to complex
```

```python
complex (4.5)      # convert float to complex
```

```python
complex (10,4.5)   # convert 2 params to complex
```

```python
complex (10,4.5,6) # can't pass 3 parameters
```

```python
complex (10,20)    # convert int to complex
```

```python
complex (2.3,4.5)  # convert float to complex
```

```python
complex (True)     # convert boolean to complex
```

```python
complex (False)    # convert boolean to complex
```

```python
complex (True,True)   # convert boolean to complex
```

```python
complex (True,False)  # convert boolean to complex
```

```python
complex (False,False) # convert boolean to complex
```

```python
complex (False,True)  # convert boolean to complex
```

```python
complex ('10')    # convert number string to complex
```

```python
complex ('10','20')    # can't convert 2 args number string to complex
```

```python
complex (10,'20')      # can't convert 2 args number string to complex
```

<h2 style="color:darkred">5. Convert all data types to string</h2>

1. In type casting to string we can't pass 2 arguments or more
2. We can type cast all to string types

```python
str (20)    # convert int to string
```

```python
str (23.4)  # convert float to string
```

```python
str (True)  # convert boolean to string
```

```python
str (3+4j)  # convert complex to string
```
