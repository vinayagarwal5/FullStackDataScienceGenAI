<h2 style="color:darkred">1. Indexing</h2>

### ➡️ Forward Indexing (Positive)

- Starts from **0** for the first character  
- Moves **left → right**

- **Start:** Index `0` → first character  
- **End:** Index `length - 1` → last character  

---

### ⬅️ Backward Indexing (Negative)

- Starts from **-1** for the last character  
- Moves **right → left**

- **Start:** Index `-1` → last character  
- **End:** Index `-length` → first character  

---

<h2 style="color:darkred">2. Slicing</h2>

<div style="font-size:16px;"></div>The most common way to slice a string is using bracket notation: <span style="color:green">string[start:stop:step]</span></div>

<table style="width:80%; font-size:14px; border-collapse:collapse; margin:auto;"> <thead> <tr style="background-color:#f2f2f2;"> <th style="color:purple; border:1px solid #333; padding:10px;">Parameter</th> <th style="color:purple; border:1px solid #333; padding:10px;">Explanation</th> </tr> </thead> <tbody> <tr> <td style="border:1px solid #333; padding:8px;"><b>start</b></td> <td style="border:1px solid #333; padding:8px;"> The index where the slice begins (<b>inclusive</b>). Default = 0 </td> </tr> <tr> <td style="border:1px solid #333; padding:8px;"><b>stop</b></td> <td style="border:1px solid #333; padding:8px;"> The index where the slice ends (<b>exclusive</b>) </td> </tr> <tr> <td style="border:1px solid #333; padding:8px;"><b>step</b></td> <td style="border:1px solid #333; padding:8px;"> (Optional) Interval between characters. Default = 1 </td> </tr> </tbody> </table>

````python
data_python = "welcome to python programming"
````

````python
data_python[:]           # Return all characters
````

````python
data_python[1:6]         # Return 1 to 5 characters
````

````python
data_python[8:]          # Return 8 to end characters
````

````python
data_python[:10]         # Return start to 9 characters
````

````python
data_python[:5]          # Return start to 4 characters
````

````python
data_python[-1:]         # Return last character
````

````python
data_python[-6:]         # Return last 6th character to end
````

````python
data_python[-7:-2]       # Return last 7th character to 3rd characterd from last
````

````python
data_python[1000:-2]     # Returns empty since so index of 1000
````

````python
data_python[10000:]     # Returns empty since so index of 10000
````

````python
data_python[0:1000]     # Returns all since stop index till 1000
````

````python
data_python[0:10:2]     # Returns step slicing returns 0,2,4,6,8
````

````python
data_python[5::2]       # Returns step slicing returns 5,7,9,... to end
````
