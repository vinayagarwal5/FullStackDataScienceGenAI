<h3 style="color:purple">4. Logical Operators</h3>
Used to combine conditional statements.

<table style="width:80%; font-size:16px; border-collapse:collapse; margin:auto; text-align:center;">
    <thead>
        <tr style="background-color:#f2f2f2;">
            <th style="color:purple; border:1px solid #333; padding:10px;">Operator</th>
            <th style="color:purple; border:1px solid #333; padding:10px;">Description</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td style="border:1px solid #333; padding:8px;"><b>and</b></td>
            <td style="border:1px solid #333; padding:8px;">Returns True if both conditions are true</td>
        </tr>
        <tr>
            <td style="border:1px solid #333; padding:8px;"><b>or</b></td>
            <td style="border:1px solid #333; padding:8px;">Returns True if at least one condition is true</td>
        </tr>
        <tr>
            <td style="border:1px solid #333; padding:8px;"><b>not</b></td>
            <td style="border:1px solid #333; padding:8px;">Reverses the result</td>
        </tr>
    </tbody>
</table>

---

<h4 style="color:black">🔍 Examples</h4>

```python
x = 7

print(x > 5 and x < 10)   # True
print(x < 5 or x > 10)    # False
print(not(x > 5))         # False
